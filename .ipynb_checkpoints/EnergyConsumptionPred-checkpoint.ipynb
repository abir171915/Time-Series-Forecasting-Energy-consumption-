```python
df["y_naive"] = df["y"].shift(48)
```

```python
test_day = "2023-09-30"
mask_test = df["ds"].dt.date == pd.to_datetime(test_day).date()
df_test = df[mask_test]  # y = actual, y_naive = same time yesterday
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

y_true = df_test["y"]
y_naive = df_test["y_naive"]

mae_naive = mean_absolute_error(y_true, y_naive)
rmse_naive = np.sqrt(mean_squared_error(y_true, y_naive))

print("Naive MAE:", mae_naive)
print("Naive RMSE:", rmse_naive)
```

```python
df_test_sorted = df_test.sort_values("ds")
fcst_small = prediction[["ds", "yhat"]]

df_eval = df_test_sorted[["ds", "y"]].merge(fcst_small, on="ds", how="inner").sort_values("ds")
print(df_eval.shape)
```

```python
y_true = df_eval["y"].values
y_prophet = df_eval["yhat"].values

mae_prophet = mean_absolute_error(y_true, y_prophet)
rmse_prophet = np.sqrt(mean_squared_error(y_true, y_prophet))

print("Prophet MAE:", mae_prophet)
print("Prophet RMSE:", rmse_prophet)
```

```python
tariff = pd.read_csv('assessment_tariff_data.csv')
```

```python
tariff.head()
```

```python
tariff['period_starting'] = pd.to_datetime(tariff['period_starting'])
```

```python
tariff.columns = ['ds', 'price']
```

```python
prediction = prediction[prediction['ds'] > df['ds'].max()][['ds','yhat']].copy()
prediction = prediction.sort_values('ds')
```

```python
plan = prediction.merge(tariff[['ds', 'price']], on = 'ds')
```

```python
plan = plan.rename(columns = {'yhat' : 'demand_kw'})
```

```python
plan.head()
```

```python
dt_hours = 0.5
plan['demand_kw'] = plan['demand_kw'] * dt_hours
```

```python
plan.head()
```

```python
plan['cost_without_batt'] = plan['demand_kw'] * plan['price']
```

```python
total_cost_without_batt = plan['cost_without_batt'].sum()
```

```python
print(f"Cost without Battery:", total_cost_without_batt)
```

```python
max_kw = 220
max_kwh = 440
dt_hours = 0.5
max_enery_given_per_period = max_kw * dt_hours
```

```python
state_kwh = 220 #start with half full
```

```python
state_kwh_list = []
charge = []
discharge = []
import_grid = []
cost_with_batt = []
```

```python
low_threshold = plan['price'].quantile(0.3)
upper_threshold = plan['price'].quantile(0.7)
```

```python
for i in range(len(plan)):
    demand_kwh = plan.iloc[i]['demand_kw']
    price = plan.iloc[i]['price']

    charge_kwh = 0.0
    discharge_kwh = 0.0

    if price <= low_threshold and state_kwh < max_kwh:
        room_kwh = max_kwh - state_kwh
        charge_kwh = min(max_enery_given_per_period, room_kwh)

    elif price >= upper_threshold and state_kwh > 0:
        max_discharge_possible = min(max_enery_given_per_period, state_kwh, demand_kwh)
        discharge_kwh = max_discharge_possible

    state_kwh = state_kwh + charge_kwh - discharge_kwh
    state_kwh = max(0.0, min(max_kwh, state_kwh), state_kwh)
    state_kwh_list.append(state_kwh)

    grid_import_kwh = demand_kwh + charge_kwh - discharge_kwh
    step_cost = grid_import_kwh * price

    state_kwh_list.append(state_kwh)
    charge.append(charge_kwh)
    discharge.append(discharge_kwh)
    import_grid.append(grid_import_kwh)
    cost_with_batt.append(step_cost)

plan['state_kwh'] = state_kwh_list
plan['charge_kwh'] = charge
plan['discharge_kwh'] = discharge
plan['grid_import_kwh'] = import_grid
plan['cost_with_batt'] = cost_with_batt

total_cost_with_batt = plan['cost_with_batt'].sum()
print(f"Cost with Battery: ",total_cost_with_batt)
```

```python
plt.figure(figsize=(10,4))
plt.plot(plan["ds"], plan["state_kwh"])
plt.xlabel("Time")
plt.ylabel("Battery SoE (kWh)")
plt.title("Battery state of energy over next 24 hours")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
